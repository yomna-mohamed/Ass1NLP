```python
from __future__ import division
import nltk, re, pprint
import urllib
from urllib import request
import pandas as pd
import random


#---------------- -------Fuction that clean the book-----------------------------------------------------------------------------------------------------
def clean_string(book):
    # Remove HTML entities
    cleanbook = re.sub(r'\&\w*;', '', book)
    # Remove @user
    cleanbook = re.sub(r'@(\w+)', '', cleanbook)
    # Remove links
    cleanbook = re.sub(r'(http|https|ftp)://[a-zA-Z0-9\\./]+', '', cleanbook)
    # Lowercase
    cleanbook = cleanbook.lower()
    # Remove Hashtags
    cleanbook= re.sub(r'#(\w+)', '',cleanbook)
    # Remove repeating chars
    cleanbook = re.sub(r'(.)\1{1,}', r'\1\1',cleanbook)
    # Remove anything that is not letters
    cleanbook = re.sub("[^a-zA-Z]", " ",cleanbook)
    # Remove anything that is less than two characters
    cleanbook = re.sub(r'\b\w{1,2}\b', '',cleanbook)
    # Remove multiple spaces
    cleanbook= re.sub(r'\s\s+', ' ', cleanbook)
    return cleanbook
#----------------------------Fuction that return a 200 random partition from the book-----------------------------------------------------------------------

def RandomDF(partitionedbook, bookName):
   df=random.choices(partitionedbook,k=200)
   #print(partitionedbook)
   df1=pd.DataFrame()
   df1['partion']= df
   df1['label']= bookName
   return df1


#------------------------------Function that patitioning the book , each partition contain a 100 word-----------------------------------------------------------------
def partitioning(cleandbook):
  list_book=[]
  count=0
  for i in range(0,200):
    list_book.append(cleandbook[count:count+100])
    count+=1
  return list_book
#----------------------------Genral Function---------------------------------------------------------------------------------------------------------------
def genralFunction(book,bookName):
  cleanbook=clean_string(book)
  #display(cleanbook)
  list_book = list(cleanbook.split(" "))
  #print(list_book)
  partitionedbook =partitioning(list_book)
  df = RandomDF(partitionedbook,bookName)
  return df


# -------------------------------------------- input data from user----------------------------------------------------------------------------------------
print("Hello ,I suggest some books links if you interested choose from these book , if not ignore them and enter your link ")
li=["https://www.gutenberg.org/cache/epub/68085/pg68085.txt","https://www.gutenberg.org/cache/epub/68084/pg68084.txt","https://www.gutenberg.org/cache/epub/68076/pg68076.txt","https://www.gutenberg.org/cache/epub/68083/pg68083.txt","https://www.gutenberg.org/cache/epub/68082/pg68082.txt","https://www.gutenberg.org/cache/epub/68081/pg68081.txt"]
print(li)
print("enter the number of books")
n = int(input())
gernral_dataFrame=pd.DataFrame()
for i in range(0,n) :
  print('Enter Book link :')
  url = input()
  bookName = url.split('/')[-1]
  #print(bookName)
  raw =  urllib.request.urlopen(url).read()
  book= raw.decode('UTF-8')
  gernral_dataFrame = gernral_dataFrame.append(genralFunction(book,bookName),ignore_index=True)

display(gernral_dataFrame.sample(len(gernral_dataFrame)))
```
